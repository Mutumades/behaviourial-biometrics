## üìå Problem Statement

### ‚ö†Ô∏è The Rising Fraud Crisis in Kenya
Mobile banking has become the backbone of finance in powering everyday transactions on platforms like KCB Bank, Equity Bank, Co-operative Bank of Kenya, Absa Bank Kenya, Stanbic Bank Kenya, Standard Chartered Kenya and Safaricom‚Äôs M-Pesa.:

However, this convenience comes with rising risk.  
Kenyan banks reported **Sh1.59 billion in cyber fraud losses in 2024‚Äì2025**, with fraud attacks becoming **more -powered, automated, and invisible**.  
Fraudsters now use **deepfakes, remote access tools (RATs), and bots** to carry out attacks faster, at scale, and without being detected.

---

### üõë Why Traditional Security Fails
Current bank security focuses only on the **login moment**:

- **PINs & Passwords** ‚Üí can be stolen, guessed, or phished  
- **One-Time Passwords (OTPs)** ‚Üí can be intercepted through SIM-swap or social engineering  
- **Physical Biometrics (Face/Fingerprint)** ‚Üí can be spoofed with photos, silicone fingerprints, or stolen templates

Once the attacker logs in, **banks have no way to know if it‚Äôs still the real user using the app.**  
This is the biggest security gap ‚Äî and why post-login fraud is exploding.

---

### üíÄ Common Post-Login Fraud in Kenya

| Fraud Type                     | Description                                                       | Real-World Risk                                                              |
|--------------------------------|-------------------------------------------------------------------|------------------------------------------------------------------------------|
| **SIM-Swap Fraud**              | Criminal hijacks the victim‚Äôs phone number via SIM replacement     | Lets attackers intercept OTPs and fully take over accounts                   |
| **Remote Access Fraud (RAT)**   | Fraudster remotely controls the victim‚Äôs phone                       | Victim unknowingly watches their phone drain their funds                     |
| **Social Engineering Scams**    | Victims are tricked into sending money to fraudsters                 | Looks legitimate to the bank because the user performs the action            |
| **Bot-Based Attacks**            | Scripts mimic human behavior to steal funds fast                     | Enables large-scale account takeovers in seconds                             |
| **High-Value Mule Transactions** | Fraudsters use compromised accounts to move large sums                  | Launders money and bypasses anti-fraud thresholds                            |

These attacks cause **huge losses and destroy public trust in mobile banking.**

# üìã Signal Groups Used in the Dataset

This dataset for the  covers five main feature groups commonly used in real-world mobile banking fraud detection systems

---

## üñêÔ∏è Touch Behavior

**What it includes:**  
How the user taps and swipes ‚Äî **tap pressure, swipe speed, swipe path, and tap duration**.

**Why it matters:**  
Everyone touches their screen differently. This is **hard to fake**, so it helps reliably identify real users.

---

## ‚å®Ô∏è Keystroke Behavior

**What it includes:**  
How the user types ‚Äî **key hold time (dwell), inter-key delay (flight), typing errors, and backspace usage**.

**Why it matters:**  
Typing rhythm is **unique to each person** and usually **changes if someone else is typing** (like a fraudster) or if the user is stressed.

---

## üß≠ Navigation Behavior

**What it includes:**  
How the user moves through the app ‚Äî **which screens they open, how long they stay, and if they use shortcuts**.

**Why it matters:**  
Fraudsters often **rush, skip normal steps, or open unusual screens** that real users don‚Äôt usually visit.

---

## üìç Context Information

**What it includes:**  
Where and when the user logs in ‚Äî **GPS city, time of day, device model, SIM card ID, and network type

**Why it matters:**  
**Big changes in location, device, or SIM card** can signal that a fraudster is using the account.

---

## üí∞ Transaction Details

**What it includes:**  
What the user does with money ‚Äî **transfer amounts and if they add a new beneficiary**.

**Why it matters:**  
Fraud often involves **very large amounts** or **quickly adding new payees**.


## üìå Basis for Choosing the Signals

I selected the features for the synthetic dataset based on **five well-known industry metrics** commonly used in behavioral biometrics.  
These metrics helped me make sure each feature is **practical (can be collected from phones), stable (consistent over time), unique (separates users), robust (hard to fake), and universal (available for all users)** ‚Äî which is important for real-world mobile banking

---

## üìä Signal Selection Metrics
| Metric               | Meaning                                                   | Example                                                                                                                                                     |
|-----------------------|------------------------------------------------------------|--------------------------------------------------------------------------------------------------------------------------------------------------------------|
| Collectability (C)    | How easily it can be collected on normal phones.           | A banking app uses Tap Pressure. This has a high collectability score of 5 because a user's tap pressure is automatically captured by a smartphone's sensors without any extra effort. |
| Stability (S)         | How consistent it is for the same user over time.           | The app learns your Preferred Shortcuts. Your habit of using a specific icon to jump to a page is a stable pattern with a high score of 5, which the app can rely on to confirm your identity. |
| Discriminability (D)   | How unique it is between different users.                   | The banking app analyzes a user's Backspace/Error Patterns. These have a high discriminability score of 5 because a person's mistakes and how they correct them are very individual and hard for a fraudster to mimic. |
| Robustness (R)         | How resistant it is to noise, SIM/device change, or spoofing.| The app uses SIM + Network Patterns. This has a high robustness score of 5 because a sudden network jump, like a criminal logging in from a new location, is a strong fraud signal that is resistant to spoofing. |
| Universality (U)       | How common it is across all users/devices.                  | Key hold time has a high universality score of 5. Every user who types a password must press and hold keys, making this a universal signal that can be collected for any user. |

---


## üìã Signal Evaluation Using C / S / D / R / U

Each feature was scored from 1 (low) to 5 (high) based on:

- **C (Collectability)** ‚Äî How easily it can be collected on normal phones  
- **S (Stability)** ‚Äî How consistent it is for the same user over time  
- **D (Discriminability)** ‚Äî How unique it is between different users  
- **R (Robustness)** ‚Äî How resistant it is to noise, SIM/device change, or spoofing  
- **U (Universality)** ‚Äî How common it is across all users/devices

```python
#  RAW Behavioral Biometrics Data Generator
# - Users: 700
# - Sessions per user: heavy-tailed (1‚Äì200), mean ~25‚Äì45 (varies)
# - Fraud ratio: ~5%
# - Fraud styles: sim_swap, remote_control, social_engineering, bot_like, high_value
# - Realism upgrades: uneven activity, human drift/quirks, correlated sensor noise,
#   subtle fraud overlap, rare edge-case frauds
# - Outputs: CSV + JSONL (collection-stage raw data with nested events)

import numpy as np
import pandas as pd
import json, hashlib, random
from datetime import datetime, timedelta
from IPython.display import display

rng = np.random.default_rng(20250911)

#  Config
N_USERS = 700
FRAUD_RATIO = 0.05  # 5%
START = datetime(2025, 6, 1, 6, 0, 0)
END   = datetime(2025, 9, 10, 23, 0, 0)

# Sessions per user: heavy-tailed (Zipf/Pareto-like)
def sample_sessions_per_user(n_users):
    a = 1.5
    raw = (rng.pareto(a, size=n_users) + 1.0) * 10  # scale controls mean
    return np.clip(raw.astype(int), 1, 200)

sessions_per_user = sample_sessions_per_user(N_USERS)

# ---------------- Helpers ----------------
def rand_ts():
    delta = END - START
    s = int(rng.integers(0, int(delta.total_seconds())))
    return (START + timedelta(seconds=s))

def hash_str(s: str) -> str:
    return hashlib.sha256(s.encode("utf-8")).hexdigest()[:16]

def maybe_missing(val, p=0.05):
    return None if rng.random() < p else val

#  Catalogs
device_models = ["Samsung A52","Samsung S21","Tecno Spark 10","Infinix Note 12","Nokia G21","iPhone 11","iPhone 12","iPhone 13"]
os_versions   = ["Android 12","Android 13","Android 14","iOS 16","iOS 17"]
resolutions   = ["720x1600","1080x2400","1440x3200","828x1792","1170x2532","1284x2778"]
carriers      = ["Safaricom","Airtel","Telkom"]
network_types = ["LTE","4G","3G","5G","WiFi"]
cities = [
    ("Nairobi",-1.2921,36.8219), ("Mombasa",-4.0435,39.6682), ("Kisumu",-0.0917,34.7680),
    ("Nakuru",-0.3031,36.0800), ("Eldoret",0.5204,35.2698), ("Thika",-1.0415,37.0827),
    ("Nyeri",-0.4201,36.9476), ("Malindi",-3.2179,40.1169), ("Machakos",-1.5177,37.2634),
    ("Meru",0.0473,37.6559),
]
screens = ["splash","login","home","balance","transactions","transfers","beneficiaries","paybills","airtime","settings","help","loans","cards"]
FRAUD_STYLES = ["sim_swap","remote_control","social_engineering","bot_like","high_value"]

def pick_fraud_style():
    return rng.choice(FRAUD_STYLES, p=[0.25,0.25,0.20,0.15,0.15])

#  User Profile (with drift anchors)
def user_profile(uid):
    city, lat, lon = cities[int(rng.integers(0, len(cities)))]
    device = rng.choice(device_models)
    osver  = rng.choice(os_versions)
    res    = rng.choice(resolutions)
    carrier = rng.choice(carriers, p=[0.6,0.32,0.08])
    sim_serial = hash_str(f"{uid}-{rng.integers(10**10, 10**12)}")
    wifi = rng.choice(["Home_WiFi","OfficeNet","Cafe_Kitengela","Uni_Guest","--"])
    # typing/touch baseline
    base_hold = float(np.clip(rng.normal(160, 40), 60, 400))
    base_gap  = float(np.clip(rng.normal(110, 35), 30, 400))
    base_pressure = float(np.clip(rng.normal(0.78, 0.12), 0.2, 1.2))
    # transactions baseline
    tx_mu = float(np.clip(abs(rng.normal(2500, 2000)), 200, 50000))
    tx_sigma = float(np.clip(abs(rng.normal(1200, 1000)), 200, tx_mu/2))
    # drift checkpoints
    move_home_after = int(rng.integers(15, 120))
    upgrade_after   = int(rng.integers(20, 150))
    return dict(home_city=city, home_lat=lat, home_lon=lon, device=device, osver=osver, res=res,
                carrier=carrier, sim_serial=sim_serial, wifi=wifi,
                base_hold=base_hold, base_gap=base_gap, base_pressure=base_pressure,
                tx_mu=tx_mu, tx_sigma=tx_sigma,
                move_home_after=move_home_after, upgrade_after=upgrade_after)

# Context effects (human quirks & correlated noise)
def session_context_effects(hour, dow, network_type, roaming):
    # Human quirks
    is_night   = (hour >= 22 or hour <= 5)
    is_weekend = (dow in ["Saturday","Sunday"])
    poor_signal = (network_type in ["3G"] or roaming or (network_type=="LTE" and rng.random()<0.1))

    hold_mult = 1.0 + (0.15 if is_night else 0.05 if is_weekend else 0.0)
    gap_mult  = 1.0 + (0.10 if is_night else 0.03 if is_weekend else 0.0)
    backspace_boost = 1 if is_night else (1 if is_weekend and rng.random()<0.3 else 0)

    # Correlated GPS noise
    gps_acc_extra = 15 if poor_signal else 0
    gps_jitter_sd = 0.02 if poor_signal else 0.003

    # Shaky hand ‚Üí lower pressure + more XY jitter
    shaky = rng.random() < (0.12 if is_night else 0.06)
    pressure_shift = -0.12 if shaky else 0.0
    tap_xy_jitter = 3 if shaky else 1

    return dict(hold_mult=hold_mult, gap_mult=gap_mult, backspace_boost=backspace_boost,
                gps_acc_extra=gps_acc_extra, gps_jitter_sd=gps_jitter_sd,
                pressure_shift=pressure_shift, tap_xy_jitter=tap_xy_jitter)

#  Event samplers
def sample_tap_events(base_pressure, pressure_shift=0.0, tap_xy_jitter=1):
    n = int(rng.integers(5, 40))
    events = []
    t0 = rng.integers(1_000_000, 2_000_000)
    for _ in range(n):
        x = int(np.clip(int(rng.integers(0, 1080)) + rng.normal(0, tap_xy_jitter), 0, 1080))
        y = int(np.clip(int(rng.integers(0, 2400)) + rng.normal(0, tap_xy_jitter), 0, 2400))
        down = int(t0 + rng.integers(0, 60_000))
        dur  = max(5, int(rng.normal(180, 60)))
        up   = down + dur
        if rng.random() > 0.25:
            pressure = float(np.clip(rng.normal(base_pressure + pressure_shift, 0.12), 0.05, 1.2))
        else:
            pressure = None
        if rng.random() < 0.01 and pressure is not None:
            pressure = float(np.clip(rng.normal(1.2, 0.2), 0.0, 2.0))
        events.append({"x": x, "y": y, "down_ms": down, "up_ms": up, "pressure": pressure})
    return events

def sample_swipe_path():
    if rng.random() < 0.15:
        return []
    k = int(rng.integers(6, 40))
    x, y = int(rng.integers(50, 1000)), int(rng.integers(100, 2200))
    t = int(rng.integers(1_000_000, 2_000_000))
    path = []
    dx = int(rng.integers(-5, 5))
    dy = int(rng.integers(10, 35))
    for _ in range(k):
        x = int(np.clip(x + dx + rng.integers(-2, 3) + rng.normal(0, 0.5), 0, 1080))
        y = int(np.clip(y + dy + rng.integers(-2, 3) + rng.normal(0, 0.5), 0, 2400))
        t += int(rng.integers(8, 25))
        path.append({"x": x, "y": y, "t": t})
    if rng.random() < 0.05:
        return []
    return path

def sample_keystrokes(base_hold, base_gap, hold_mult=1.0, gap_mult=1.0, bot_like=False):
    n = int(rng.integers(5, 40))
    seq = []
    base = int(rng.integers(1_000_000, 2_000_000))
    for _ in range(n):
        c = random.choice(list("abcdefghijklmnopqrstuvwxyz0123456789"))
        down = base + int(rng.integers(10, 70))
        if bot_like:
            hold = int(max(10, rng.normal(60, 10)))
            gap  = int(max(5, rng.normal(40, 10)))
        else:
            hold = int(max(20, rng.normal(base_hold*hold_mult, 45)))
            gap  = int(max(10, rng.normal(base_gap*gap_mult, 35)))
        up = down + hold
        seq.append({"key": c, "down_ms": down, "up_ms": up})
        base = up + gap
    backspaces = int(max(0, rng.integers(0, 4))) + (1 if hold_mult>1.05 else 0)
    return seq, backspaces

def sample_navigation(remote_control=False, social_engineering=False):
    flow = ["splash","login","home"]
    times = [round(float(np.clip(rng.normal(2.5, 1.2), 0.2, 25.0)), 2) for _ in flow]
    clicks = []
    shortcut_used = bool(rng.random() < 0.7)

    if social_engineering:
        extra_screens = ["transfers","beneficiaries"]
        flow += extra_screens
        times += [round(float(np.clip(rng.normal(1.0, 0.5), 0.1, 10.0)), 2) for _ in extra_screens]
        clicks += [f"btn_{s}" for s in extra_screens]
        shortcut_used = False
    else:
        extra = int(rng.integers(1, 5))
        chosen = list(rng.choice(screens[3:], size=extra, replace=False))
        flow += chosen
        times += [round(float(np.clip(rng.normal(2.0, 1.0), 0.2, 20.0)), 2) for _ in chosen]
        clicks += [f"btn_{s}" for s in chosen if s not in ("splash","home")]

    if remote_control and rng.random() < 0.7:
        flow = ["splash","login","home","help","settings","transfers"]
        times = [round(float(np.clip(rng.normal(1.5, 0.8), 0.2, 20.0)), 2) for _ in flow]
        clicks = [f"btn_{s}" for s in flow if s not in ("splash","home")]
        shortcut_used = False

    return flow, times, clicks, shortcut_used

#  Main generation
rows = []
edgecases_remaining = 20  # rare remote-control-like sessions

for i in range(N_USERS):
    uid = f"U{str(i+1).zfill(4)}"
    prof = user_profile(uid)
    base_device_id = hash_str(uid + prof["device"] + prof["osver"])
    n_sess = sessions_per_user[i]

    for s in range(n_sess):
        ts = rand_ts()
        dow = ts.strftime("%A")
        hour = ts.hour

        network_type = rng.choice(network_types, p=[0.35,0.25,0.10,0.15,0.15])  # bias to LTE/4G
        roaming = bool(rng.random() < 0.05)
        ctx = session_context_effects(hour, dow, network_type, roaming)

        # Fraud assignment
        is_fraud = (rng.random() < FRAUD_RATIO)
        fraud_style = pick_fraud_style() if is_fraud else None

        # Rare edge-case frauds
        if edgecases_remaining > 0 and rng.random() < 0.0008:
            is_fraud = True
            fraud_style = "remote_control"
            edgecases_remaining -= 1

        # Geo (home vs new city; correlated jitter/accuracy)
        if s >= prof["move_home_after"] and rng.random() < 0.30:
            city, lat, lon = random.choice(cities)
        else:
            if is_fraud and fraud_style in ("sim_swap","high_value") and rng.random() < 0.7:
                city, lat, lon = random.choice(cities)
            else:
                if rng.random() < 0.85:
                    city, lat, lon = prof["home_city"], prof["home_lat"], prof["home_lon"]
                else:
                    city, lat, lon = random.choice(cities)
        city = maybe_missing(city, p=0.04)
        if city is None:
            lat = None; lon = None; acc_m = None
        else:
            lat = float(lat + rng.normal(0, ctx["gps_jitter_sd"]))
            lon = float(lon + rng.normal(0, ctx["gps_jitter_sd"]))
            acc_m = round(float(np.clip(rng.normal(12 + ctx["gps_acc_extra"], 10), 3, 150)), 1)

        # SIM/network/device & possible device upgrade
        # --- Context: SIM/network/device (Kenya- realistic) ---
        carrier = prof["carrier"]; sim_serial = prof["sim_serial"]
        device_model = prof["device"]; osver = prof["osver"]; res = prof["res"]
        device_id = base_device_id
        ip = f"{rng.integers(1,255)}.{rng.integers(0,255)}.{rng.integers(0,255)}.{rng.integers(0,255)}"

        # In Kenya most users use mobile data, not Wi-Fi.
        # Make Wi-Fi SSID rare (~15% present, ~85% missing).
        wifi_ssid = prof["wifi"] if rng.random() < 0.15 else None

        # Possible device upgrade over time
        if s >= prof["upgrade_after"] and rng.random() < 0.4:
            device_model = rng.choice(device_models)
            osver = rng.choice(os_versions)
            device_id = hash_str(uid + str(rng.integers(10**6,10**8)))

        # Fraud effects on SIM/Device
        if is_fraud:
            if fraud_style == "sim_swap":
                if rng.random() < 0.7:
                    sim_serial = hash_str(uid + str(rng.integers(10**9,10**11)))
                if rng.random() < 0.5:
                    carrier = rng.choice(carriers)
                if rng.random() < 0.4:
                    device_model = rng.choice(device_models)
                    osver = rng.choice(os_versions)
                    device_id = hash_str(uid + str(rng.integers(10**6,10**8)))
            elif fraud_style == "high_value":
                if rng.random() < 0.5:
                    device_model = rng.choice(device_models)
                    device_id = hash_str(uid + str(rng.integers(10**6,10**8)))

        # Missingness on context fields
        device_model = maybe_missing(device_model, p=0.06)
        osver       = maybe_missing(osver,       p=0.05)
        res         = maybe_missing(res,         p=0.03)
        carrier     = maybe_missing(carrier,     p=0.01)  # carrier is almost always present on mobile data
        # wifi_ssid already has ~85% missing by design; no extra missingness applied


        # Events
        bot_like = (is_fraud and fraud_style == "bot_like")
        taps = sample_tap_events(prof["base_pressure"], pressure_shift=ctx["pressure_shift"], tap_xy_jitter=ctx["tap_xy_jitter"])
        swipe = sample_swipe_path()
        keys, backspaces = sample_keystrokes(prof["base_hold"], prof["base_gap"], hold_mult=ctx["hold_mult"], gap_mult=ctx["gap_mult"], bot_like=bot_like)
        backspaces += ctx["backspace_boost"]

        # Navigation variants
        remote = (is_fraud and fraud_style == "remote_control")
        social = (is_fraud and fraud_style == "social_engineering")
        flow, times, clicks, shortcut = sample_navigation(remote_control=remote, social_engineering=social)

        # Subtle fraud overlap: some fraud sessions look normal
        if is_fraud and rng.random() < 0.25:
            keys, backspaces = sample_keystrokes(prof["base_hold"], prof["base_gap"], hold_mult=1.0, gap_mult=1.0, bot_like=False)
            flow, times, clicks, shortcut = sample_navigation(remote_control=False, social_engineering=False)

        # Transactions
        amount = abs(rng.normal(prof["tx_mu"], prof["tx_sigma"]))
        if is_fraud and fraud_style in ("high_value","sim_swap","social_engineering"):
            amount *= rng.uniform(2.0, 8.0)
        new_beneficiary = 1 if (is_fraud and rng.random() < 0.6) else (1 if rng.random() < 0.05 else 0)

        # Outliers
        if rng.random() < 0.004:
            amount *= 10
        if rng.random() < 0.004 and (city is not None):
            lat = float(lat + rng.normal(0, 0.25))
            lon = float(lon + rng.normal(0, 0.25))

        rows.append({
            "user_id": uid,
            "session_id": f"{uid}_S{s+1:03d}",
            "timestamp": ts.isoformat(),
            "day_of_week": dow,
            "hour_of_day": hour,
            "tap_events_json": json.dumps(taps, ensure_ascii=False),
            "swipe_path_json": json.dumps(swipe, ensure_ascii=False) if len(swipe)>0 else None,
            "keystroke_events_json": json.dumps(keys, ensure_ascii=False),
            "backspace_count": int(backspaces),
            "screen_sequence_json": json.dumps(flow, ensure_ascii=False),
            "time_on_screen_json": json.dumps(times, ensure_ascii=False),
            "button_clicks_json": json.dumps(clicks, ensure_ascii=False),
            "shortcut_used": bool(shortcut),
            "latitude": None if city is None else float(lat),
            "longitude": None if city is None else float(lon),
            "accuracy_m": None if city is None else float(np.clip(acc_m, 1, 200)),
            "city": city,
            "country": "KE",
            "carrier_name": carrier,
            "sim_serial_hashed": sim_serial,
            "network_type": network_type,
            "roaming": bool(roaming),
            "ip_address": ip,
            "wifi_ssid": wifi_ssid,
            "device_model": device_model,
            "os_version": osver,
            "screen_resolution": res,
            "device_id_hashed": device_id,
            "transfer_amount": round(float(amount), 2),
            "new_beneficiary": int(new_beneficiary),
            "label": 0 if is_fraud else 1,
            "fraud_style": fraud_style
        })

raw_v4 = pd.DataFrame(rows)

#  Save
csv_path = "behavioral_biometrics_RAW_v4.csv"
jsonl_path = "behavioral_biometrics_RAW_v4.jsonl"
df_out = raw_v4.where(pd.notna(raw_v4), None)  # ensure JSONL has null, not NaN
df_out.to_csv(csv_path, index=False)
df_out.to_json(jsonl_path, orient="records", lines=True, force_ascii=False)

# Quick summary & preview
summary = {
    "rows": len(df_out),
    "users": df_out["user_id"].nunique(),
    "fraud_rate_observed": float((df_out["label"]==0).mean()),
    "fraud_style_counts": df_out["fraud_style"].value_counts(dropna=False).to_dict(),
    "sessions_per_user_stats": {
        "min": int(pd.Series(sessions_per_user).min()),
        "max": int(pd.Series(sessions_per_user).max()),
        "mean": float(pd.Series(sessions_per_user).mean()),
        "median": float(pd.Series(sessions_per_user).median())
    }
}
print("Summary:", json.dumps(summary, indent=2))
display(df_out.head(10))
print("Saved:", csv_path, "and", jsonl_path)
```

5.0 Load Raw Data

This part loads the synthetic behavioral sessions and converts timestamps into the correct format.
It provides a clean starting dataset for further processing.

```python
import json, math
import numpy as np
import pandas as pd
from datetime import datetime

from sklearn.model_selection import GroupShuffleSplit
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import VarianceThreshold

raw = pd.read_csv("behavioral_biometrics_RAW_v4.csv", low_memory=False)
raw["timestamp"] = pd.to_datetime(raw["timestamp"], errors="coerce")
```

5.1    **Feature Engineering (from raw events to numbers)**

The code below converts raw app events into numeric features that describe user behavior, such as:

Touch: tap counts, durations, and pressure

Swipes: speed, variation, and curvature

Typing: key hold times, typing speed, and mistakes

Navigation: number of screens, clicks, and time to reach the ‚ÄúTransfers‚Äù screen

These features represent how a user interacts with the app (not what they type).

```python
def safe_load_json(x):
    if pd.isna(x) or x is None or x == "": return []
    try: return json.loads(x)
    except: return []

def summarize_taps(taps):
    n = len(taps); durs=[]; prs=[]
    for e in taps:
        durs.append(max((e.get("up_ms",0)-e.get("down_ms",0)),0))
        p = e.get("pressure")
        if p is not None: prs.append(p)
    return dict(tap_count=n,
                tap_dur_mean=np.mean(durs) if durs else np.nan,
                tap_dur_std =np.std(durs)  if durs else np.nan,
                press_mean  =np.mean(prs)  if prs  else np.nan,
                press_std   =np.std(prs)   if prs  else np.nan)

def summarize_swipe(path):
    k=len(path)
    if k<2:
        return dict(swipe_points=k, swipe_speed_mean=np.nan, swipe_speed_std=np.nan, swipe_curviness=np.nan)
    speeds=[]; curv=0.0
    for i in range(1,k):
        x1,y1,t1 = path[i-1]["x"], path[i-1]["y"], path[i-1]["t"]
        x2,y2,t2 = path[i]["x"],  path[i]["y"],  path[i]["t"]
        dt=max(t2-t1,1); dist=math.hypot(x2-x1,y2-y1)
        speeds.append(dist/dt)
        if i>=2:
            x0,y0,_ = path[i-2]["x"], path[i-2]["y"], path[i-2]["t"]
            v1=(x1-x0,y1-y0); v2=(x2-x1,y2-y1)
            dot=v1[0]*v2[0]+v1[1]*v2[1]
            n1=math.hypot(*v1); n2=math.hypot(*v2)
            if n1>0 and n2>0:
                cosang=max(min(dot/(n1*n2),1),-1)
                curv+=math.acos(cosang)
    return dict(swipe_points=k,
                swipe_speed_mean=np.mean(speeds),
                swipe_speed_std =np.std(speeds),
                swipe_curviness =curv/k)

def summarize_keystrokes(keys):
    n=len(keys); holds=[]; gaps=[]
    for i,e in enumerate(keys):
        d=e.get("down_ms"); u=e.get("up_ms")
        if d is not None and u is not None: holds.append(max(u-d,0))
        if i>0:
            prev_u=keys[i-1].get("up_ms")
            if prev_u is not None and d is not None: gaps.append(max(d-prev_u,0))
    return dict(key_count=n,
                hold_mean=np.mean(holds) if holds else np.nan,
                hold_std =np.std(holds)  if holds else np.nan,
                gap_mean =np.mean(gaps)  if gaps  else np.nan,
                gap_std  =np.std(gaps)   if gaps  else np.nan)

def summarize_navigation(screens, dwell, clicks):
    n_screens = len(screens) if isinstance(screens,list) else 0
    total_time = float(np.sum(dwell)) if isinstance(dwell,list) and len(dwell)>0 else np.nan
    n_clicks = len(clicks) if isinstance(clicks,list) else 0
    time_to_transfers = np.nan
    if isinstance(screens,list) and isinstance(dwell,list) and len(screens)==len(dwell):
        cum=0.0
        for s,t in zip(screens,dwell):
            cum += float(t)
            if s=="transfers":
                time_to_transfers=cum; break
    return dict(nav_screens=n_screens, nav_total_time=total_time, nav_clicks=n_clicks,
                nav_time_to_transfers=time_to_transfers)

# Parse & summarize
raw["tap_parsed"]   = raw["tap_events_json"].map(safe_load_json)
raw["swipe_parsed"] = raw["swipe_path_json"].map(safe_load_json)
raw["keys_parsed"]  = raw["keystroke_events_json"].map(safe_load_json)
raw["scr_parsed"]   = raw["screen_sequence_json"].map(safe_load_json)
raw["dwell_parsed"] = raw["time_on_screen_json"].map(safe_load_json)
raw["clicks_parsed"]= raw["button_clicks_json"].map(safe_load_json)

tap_feats   = raw["tap_parsed"].apply(summarize_taps).apply(pd.Series)
swipe_feats = raw["swipe_parsed"].apply(summarize_swipe).apply(pd.Series)
key_feats   = raw["keys_parsed"].apply(summarize_keystrokes).apply(pd.Series)
nav_feats   = pd.DataFrame([summarize_navigation(s,t,c) for s,t,c in
                            zip(raw["scr_parsed"], raw["dwell_parsed"], raw["clicks_parsed"])])

fe = pd.concat([raw, tap_feats, swipe_feats, key_feats, nav_feats], axis=1)
```

5.2 **EDA-Driven Fixes (cleaning and context)**

The code below cleans and enhances the engineered features:

Fills missing values (e.g., sets swipe speed to 0 when no swipe data exists)

Creates context features from the previous session (city, device, SIM changes)

Calculates travel speed between sessions and flags impossible speed values

Computes each user‚Äôs amount z-score to measure how unusual a transfer amount is

These steps ensure the features are reliable and realistic before analysis.

```python
# Transfers: flag + fill time=0 when never opened
fe["did_open_transfers"] = fe["nav_time_to_transfers"].notna().astype(int)
fe["nav_time_to_transfers"] = fe["nav_time_to_transfers"].fillna(0.0)

# Swipes: if no swipe points, set derived metrics to 0
fe["swipe_points"] = fe["swipe_points"].fillna(0)
for col in ["swipe_speed_mean","swipe_speed_std","swipe_curviness"]:
    fe.loc[fe["swipe_points"]==0, col] = 0.0

# Drop leakage-y helper if present
fe = fe.drop(columns=["fraud_style"], errors="ignore")

# Sort + previous-session context
fe = fe.sort_values(["user_id","timestamp"]).reset_index(drop=True)
for col in ["city","device_id_hashed","sim_serial_hashed","latitude","longitude","timestamp"]:
    fe[f"prev_{col}"] = fe.groupby("user_id")[col].shift(1)

# Change flags
fe["city_changed"]   = (fe["city"] != fe["prev_city"]).astype(float)
fe["device_changed"] = (fe["device_id_hashed"] != fe["prev_device_id_hashed"]).astype(float)
fe["sim_changed"]    = (fe["sim_serial_hashed"] != fe["prev_sim_serial_hashed"]).astype(float)

# Travel speed (km/h)
def haversine_km(lat1,lon1,lat2,lon2):
    if any(pd.isna([lat1,lon1,lat2,lon2])): return np.nan
    R=6371.0
    p1,p2 = np.radians(lat1), np.radians(lat2)
    dphi  = np.radians(lat2-lat1); dl = np.radians(lon2-lon1)
    a = np.sin(dphi/2)**2 + np.cos(p1)*np.cos(p2)*np.sin(dl/2)**2
    return 2*R*np.arcsin(np.sqrt(a))

dist_km = fe.apply(lambda r: haversine_km(r["prev_latitude"], r["prev_longitude"],
                                          r["latitude"], r["longitude"]), axis=1)
time_h  = (fe["timestamp"] - fe["prev_timestamp"]).dt.total_seconds()/3600.0
fe["travel_speed_kmh"] = dist_km / time_h
fe.loc[(time_h<=0) | (time_h.isna()), "travel_speed_kmh"] = np.nan
fe["travel_speed_kmh"] = fe["travel_speed_kmh"].clip(0, 500).fillna(0.0)
fe["impossible_speed_flag"] = (fe["travel_speed_kmh"] > 250).astype(float)

# Per-user amount z-score (winsorized)
fe["amount_z_user"] = fe.groupby("user_id")["transfer_amount"].transform(
    lambda s: (s - s.mean()) / (s.std(ddof=0) + 1e-6)
).clip(-5,5)
```

5.3 ‚ûï Combined Features (hidden fraud patterns)

This part creates combined features that capture suspicious patterns that appear only when multiple behaviors happen together, for example:

Very fast transfers to a new beneficiary

Device and SIM changed at the same time (possible SIM-swap)

High-value transfers at night

Bot-like typing or swiping (too perfect or too straight)

A context anomaly score that sums unusual behavior changes

These combinations help increase accuracy and realism during fraud detection.       




| Combo Feature              | How to Calculate                                             | Why It Helps                                                      |
| -------------------------- | ------------------------------------------------------------ | ----------------------------------------------------------------- |
| `fast_new_transfer`        | `nav_time_to_transfers < 5` **AND** `new_beneficiary == 1`   | Fraudsters often rush to send money to a new person               |
| `device_sim_change`        | `device_changed == 1` **AND** `sim_changed == 1`             | SIM-swap attacks often involve device + SIM change together       |
| `tap_backspace_combo`      | `tap_dur_mean * backspace_count`                             | Slow taps + many backspaces means unfamiliar user typing          |
| `speed_curviness_ratio`    | `swipe_speed_mean / (swipe_curviness+1e-5)`                  | Fast perfectly-straight swipes can indicate bots                  |
| `pressure_speed_mismatch`  | `press_mean / swipe_speed_mean`                              | Very hard presses but fast swipes are unnatural                   |
| `transfer_hour_risk`       | `transfer_amount * (hour_of_day in [0‚Äì5])`                   | High amounts at odd hours = suspicious                            |
| `location_jump_speed`      | `travel_speed_kmh > 500` **AND** `city_changed == 1`         | Impossible travel = likely remote access                          |
| `low_variability_typing`   | `hold_std < 0.02` **AND** `gap_std < 0.02`                   | Bots type with perfectly steady timing                            |
| `fast_session_high_value`  | `nav_total_time < 10` **AND** `transfer_amount > user_avg*3` | Criminal quickly sends big amount                                 |
| `network_context_risk`     | `roaming == 1` **AND** `accuracy_m > 1000`                   | Roaming + inaccurate GPS = possible spoofing                      |
| `frequent_backtrack`       | `nav_clicks / nav_screens > 3`                               | Many clicks per screen = confusion = not normal user              |
| `high_pressure_low_taps`   | `press_mean > 1.5` **AND** `tap_count < 5`                   | Overly forceful but few taps = stress or automation               |
| `shortcut_absent_slow_nav` | `shortcut_used == 0` **AND** `nav_time_to_transfers > 20`    | Legit users usually use shortcuts; not doing so = unfamiliar user |
| `amount_zscore_extreme`    | `abs(amount_z_user) > 3`                                     | Amount far outside user‚Äôs normal range = risk                     |
| `contextal_anomaly_score`  | `(sim_changed + device_changed + city_changed + roaming)`    | Sum of unusual context events; higher = riskier                   |


```python
EPS = 1e-5  # tiny number for safe division

fe["fast_new_transfer"] = ((fe["nav_time_to_transfers"] < 5) & (fe["new_beneficiary"] == 1)).astype(int)
fe["device_sim_change"] = ((fe["device_changed"] == 1) & (fe["sim_changed"] == 1)).astype(int)
fe["tap_backspace_combo"] = fe["tap_dur_mean"].fillna(0) * fe["backspace_count"].fillna(0)
fe["speed_curviness_ratio"] = fe["swipe_speed_mean"].fillna(0) / (fe["swipe_curviness"].fillna(0) + EPS)
fe["pressure_speed_mismatch"] = fe["press_mean"].fillna(0) / (fe["swipe_speed_mean"].fillna(0) + EPS)
night_flag = (fe["hour_of_day"] < 6).astype(int)
fe["transfer_hour_risk"] = fe["transfer_amount"].fillna(0) * night_flag
fe["location_jump_speed"] = ((fe["travel_speed_kmh"] > 500) & (fe["city_changed"] == 1)).astype(int)
fe["low_variability_typing"] = ((fe["hold_std"] < 0.02) & (fe["gap_std"] < 0.02)).astype(int)
user_avg = fe.groupby("user_id")["transfer_amount"].transform("mean")
global_avg = fe["transfer_amount"].mean()
threshold = (user_avg.fillna(global_avg) * 3.0)
fe["fast_session_high_value"] = ((fe["nav_total_time"] < 10) & (fe["transfer_amount"] > threshold)).astype(int)
fe["network_context_risk"] = ((fe["roaming"] == 1) & (fe["accuracy_m"] > 1000)).astype(int)
fe["frequent_backtrack"] = fe["nav_clicks"].fillna(0) / (fe["nav_screens"].fillna(0) + EPS)
fe["high_pressure_low_taps"] = ((fe["press_mean"] > 1.5) & (fe["tap_count"] < 5)).astype(int)
fe["shortcut_absent_slow_nav"] = ((fe["shortcut_used"] == 0) & (fe["nav_time_to_transfers"] > 20)).astype(int)
fe["amount_zscore_extreme"] = (fe["amount_z_user"].abs() > 3).astype(int)
ctx_cols = ["sim_changed","device_changed","city_changed","roaming"]
fe["contextal_anomaly_score"] = fe[ctx_cols].fillna(0).sum(axis=1)

# (Optional) Save a v5 for EDA that includes combos
fe.to_csv("behavioral_biometrics_FEATURED_v5.csv", index=False)
print("Saved: behavioral_biometrics_FEATURED_v5.csv (with combos)")
```

5.4 Select Columns (features and categories)

This part defines which columns are numeric and which are categorical,
so that the preprocessing pipeline can correctly scale numeric values and one-hot encode categorical values.

```python
target = "label"   # 1 = legit, 0 = fraud

num_cols = [
    # touch
    "tap_count","tap_dur_mean","tap_dur_std","press_mean","press_std",
    # swipe
    "swipe_points","swipe_speed_mean","swipe_speed_std","swipe_curviness",
    # keys
    "key_count","hold_mean","hold_std","gap_mean","gap_std","backspace_count",
    # nav
    "nav_screens","nav_total_time","nav_clicks","nav_time_to_transfers","shortcut_used",
    "did_open_transfers",
    # context numeric
    "latitude","longitude","accuracy_m","hour_of_day","travel_speed_kmh",
    "amount_z_user","transfer_amount",
    # change flags
    "city_changed","device_changed","sim_changed","impossible_speed_flag",
    # extras
    "new_beneficiary","roaming",
    # ‚ûï combos
    "fast_new_transfer","device_sim_change","tap_backspace_combo","speed_curviness_ratio",
    "pressure_speed_mismatch","transfer_hour_risk","location_jump_speed","low_variability_typing",
    "fast_session_high_value","network_context_risk","frequent_backtrack","high_pressure_low_taps",
    "shortcut_absent_slow_nav","amount_zscore_extreme","contextal_anomaly_score"
]

cat_cols = ["day_of_week","city","carrier_name","network_type","wifi_ssid",
            "device_model","os_version","screen_resolution"]

X = fe[num_cols + cat_cols].copy()
y = fe[target].astype(int)
groups = fe["user_id"].values

# Drop super-sparse numeric columns (>95% missing)
miss_ratio = X[num_cols].isna().mean()
num_cols = [c for c in num_cols if miss_ratio.get(c,0.0) <= 0.95]
```

5.5 Grouped Train/Test Split (by user)

This part splits the dataset into training and testing sets by user ID.
Each user appears in only one set. This prevents the model from memorizing specific users and tests whether it can generalize to new users.

```python
gss = GroupShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))
X_train, X_test = X.iloc[train_idx].copy(), X.iloc[test_idx].copy()
y_train, y_test = y.iloc[train_idx].copy(), y.iloc[test_idx].copy()
```

5.6 Preprocess, Scale, Encode, and Save

This part builds and applies the preprocessing pipeline:

Imputes missing values

Scales numeric features

One-hot encodes categorical features

Removes zero-variance columns

Saves the processed feature matrices and feature names for later modeling

This finalizes the preprocessing stage and produces data ready for training models.

```python
num_pipe = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="median")),
    ("scale", StandardScaler()),
    ("vth", VarianceThreshold(threshold=0.0))
])

cat_pipe = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="constant", fill_value="Unknown")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

prep = ColumnTransformer(
    transformers=[
        ("num", num_pipe, num_cols),
        ("cat", cat_pipe, cat_cols),
    ],
    remainder="drop"
)

X_train_proc = prep.fit_transform(X_train)
X_test_proc  = prep.transform(X_test)

print("Train shape:", X_train_proc.shape, " Test shape:", X_test_proc.shape)

# Save artifacts
np.save("X_train_proc.npy", X_train_proc)
np.save("X_test_proc.npy",  X_test_proc)
y_train.to_numpy().dump(open("y_train.npy","wb"))
y_test.to_numpy().dump(open("y_test.npy","wb"))

# Keep a copy of feature-engineered table (v5 already saved above)
fe.to_csv("behavioral_biometrics_FEATURED_v5.csv", index=False)

# Store feature names
ohe = prep.named_transformers_["cat"].named_steps["onehot"]
cat_names = ohe.get_feature_names_out(cat_cols)
num_names_after_vth = prep.named_transformers_["num"].get_feature_names_out(num_cols)
feature_names = list(num_names_after_vth) + list(cat_names)
pd.Series(feature_names).to_csv("feature_names.csv", index=False)

print("Saved: X_train_proc.npy, X_test_proc.npy, y_train.npy, y_test.npy, feature_names.csv, behavioral_biometrics_FEATURED_v5.csv")
```

**Step 7 ‚Äî Model Training and Evaluation**

This step trains machine learning models to classify sessions as genuine (1) or fraud (0) based on behavioral biometrics features.
The goal is to test if the engineered and combined features can accurately detect fraud on unseen data.

What this step does:

1
Loads the preprocessed train/test sets

2
Trains three different models:

Logistic Regression (baseline linear model)

Random Forest (tree-based ensemble)

XGBoost (gradient boosting for higher accuracy)

3
Evaluates each model using:

Accuracy, Precision, Recall, F1-score

ROC-AUC curve

Confusion matrix

4
Prints and compares the results to pick the best model

This helps identify which algorithm works best for detecting mobile banking fraud using behavioral biometrics.

```python

# Step 7: Model Training and Evaluation
# Train and evaluate models to assess the effectiveness of features for fraud detection.
# Multiple model types are used for performance comparison.

import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

# ---------- 1) Load the processed data ----------
# Load preprocessed training and testing data saved previously.
X_train = np.load("X_train_proc.npy", allow_pickle=True)
X_test  = np.load("X_test_proc.npy", allow_pickle=True)
y_train = np.load("y_train.npy", allow_pickle=True)
y_test  = np.load("y_test.npy", allow_pickle=True)

print("Train shape:", X_train.shape, " Test shape:", X_test.shape)

# ---------- 2) Define the models ----------
# Set up models for testing. Class weights are used to handle data imbalance.
models = {
    "Logistic Regression": LogisticRegression(max_iter=200, class_weight='balanced', random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42),
    "XGBoost": XGBClassifier(n_estimators=300, learning_rate=0.1, max_depth=5, subsample=0.8,
                             colsample_bytree=0.8, eval_metric='logloss', use_label_encoder=False, random_state=42)
}

results = []

# ---------- 3) Train and Evaluate Each Model ----------
# Iterate through each model to train and evaluate performance on the test set.
for name, model in models.items():
    print(f"\nüìå Training {name} ...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1] # Probabilities obtained for ROC-AUC

    # Calculate key evaluation metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc = roc_auc_score(y_test, y_prob)

    results.append([name, acc, prec, rec, f1, roc])

    print(f"Accuracy: {acc:.3f} | Precision: {prec:.3f} | Recall: {rec:.3f} | F1: {f1:.3f} | ROC-AUC: {roc:.3f}")

    # Confusion Matrix displayed to show true positives, false positives, etc.
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fraud (0)", "Legit (1)"])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix ‚Äî {name}")
    plt.show()

# ---------- 4) Compare All Models ----------
# Results compiled into a table for easy comparison of model performance.
results_df = pd.DataFrame(results, columns=["Model","Accuracy","Precision","Recall","F1","ROC-AUC"])
print("\nüìä Model Performance Summary:")
print(results_df.sort_values(by="ROC-AUC", ascending=False).reset_index(drop=True))
```

**Model Results ‚Äî Interpretation**

The best performing model is XGBoost, with:

Accuracy: 98.6%

Precision: 98.7%

Recall: 99.8%

F1-score: 99.2%

ROC-AUC: 0.964

Random Forest also performed strongly with 97.9% accuracy and 99.9% recall.

Logistic Regression achieved lower recall (92.9%) and accuracy (92.6%), indicating it missed more fraud sessions compared to the tree-based models.

**Why this matters:**

High recall is critical in fraud detection because it ensures most fraudulent sessions are caught.

The combination of touch, swipe, typing, navigation, and context-based features contributed to strong performance across all models.

The engineered combined features (e.g., SIM + device change, location jumps, fast transfers to new beneficiaries) gave tree-based models a strong advantage.

**Real-world expectations**:

In real banking environments, similar behavioral biometrics systems often reach 85‚Äì92% accuracy and 80‚Äì90% recall on noisier data.

Achieving higher scores (as seen here) is possible when features are well-designed and sessions are clearly separated, but real-world performance may vary based on data quality and noise.

Conclusion:

XGBoost is selected as the primary model for further tuning and feature importance analysis in Step 8.

**Step 8 ‚Äî Model Tuning and Feature Importance**

This step improves the best model (XGBoost) and explains which behavioral features contribute most to fraud detection.

Goals:

Hyperparameter tuning: Search for the best settings to boost accuracy and recall.

Cross-validation: Validate performance more reliably than a single train/test split.

Feature importance analysis: Identify which features influence the model most.

This step makes the model more accurate, more stable, and more explainable.

```python

# Step 8: Model Tuning and Feature Importance

import numpy as np
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from xgboost import XGBClassifier, plot_importance
from sklearn.metrics import classification_report, roc_auc_score
import matplotlib.pyplot as plt

#  Load the processed train/test data created earlier
X_train = np.load("X_train_proc.npy", allow_pickle=True)
X_test  = np.load("X_test_proc.npy", allow_pickle=True)
y_train = np.load("y_train.npy", allow_pickle=True)
y_test  = np.load("y_test.npy", allow_pickle=True)

#  Start with a base XGBoost model (our best from Step 7)
xgb_base = XGBClassifier(
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)

#  Define a range of values for tuning
# These are common hyperparameters that affect depth, learning speed, and how many trees are used.
param_dist = {
    'n_estimators': [200, 300, 400],
    'max_depth': [3, 4, 5, 6],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'min_child_weight': [1, 3, 5]
}

#  Use 5-fold cross-validation to get reliable scores during tuning
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

#  RandomizedSearch tries different combinations and finds the best one
search = RandomizedSearchCV(
    estimator=xgb_base,
    param_distributions=param_dist,
    n_iter=20,                # try 20 random combos
    scoring='roc_auc',        # rank them by ROC-AUC
    n_jobs=-1,                 # use all cores to go faster
    cv=cv,
    verbose=1,
    random_state=42
)

print("‚è≥ Searching for best hyperparameters...")
search.fit(X_train, y_train)

print("\n‚úÖ Best Parameters found:")
print(search.best_params_)

#  Evaluate the best tuned model on the test set
best_xgb = search.best_estimator_
y_pred = best_xgb.predict(X_test)
y_prob = best_xgb.predict_proba(X_test)[:, 1]

print("\nüìä Classification Report on Test Set:")
print(classification_report(y_test, y_pred, target_names=['Fraud (0)','Legit (1)']))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

#  Plot the top 20 most important features to see what drives predictions
feature_names = pd.read_csv("feature_names.csv", header=None)[0].tolist()

plt.figure(figsize=(8,10))
plot_importance(best_xgb, max_num_features=20, height=0.6, importance_type='gain', show_values=False)
plt.title("Top 20 Most Important Features")
plt.show()
```

```python
%%writefile app.py
# app.py ‚Äî Behavioral Biometrics Fraud Dashboard (Designed)
# ------------------------------------------------------------
# Highlights:
# - Tabs for Single, Batch, Insights, Settings
# - Grouped inputs with accordions
# - KPI metrics + clean plots
# - CSV template download
# ------------------------------------------------------------

import io
import numpy as np
import pandas as pd
import joblib
import streamlit as st
import matplotlib.pyplot as plt

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, confusion_matrix, ConfusionMatrixDisplay
)

# =============================
# Page & global UI styling
# =============================
st.set_page_config(page_title="Fraud Detection Dashboard", layout="wide", page_icon="üõ°Ô∏è")

st.markdown("""
<style>
/* tighten top spacing & make headings cleaner */
.block-container { padding-top: 1.2rem; padding-bottom: 2rem; }
h1, h2, h3 { margin-bottom: 0.4rem; }
/* subtle card look for sections */
.section-card {
  padding: 1rem 1.2rem; border: 1px solid #e6e6e6; border-radius: 12px; background: #ffffff10;
}
.caption { color: #666; font-size: 0.9rem; }
.badge { display:inline-block; padding: 0.22rem 0.55rem; border-radius: 999px; border:1px solid #e6e6e6; margin-right: 0.4rem; font-size: 0.78rem; }
hr { margin: 0.8rem 0 1.2rem 0; }
</style>
""", unsafe_allow_html=True)

# Header
st.markdown("## üõ°Ô∏è Behavioral Biometrics ‚Äî Fraud Detection Dashboard")
st.markdown(
    '<span class="badge">XGBoost (tuned)</span>'
    '<span class="badge">Thresholdable Alerts</span>'
    '<span class="badge">Explainability</span>',
    unsafe_allow_html=True
)
st.write("Score mobile sessions, tune the decision threshold, and review metrics and feature importance.")

# =============================
# Load artifacts (cached)
# =============================
@st.cache_resource
def load_artifacts():
    model = joblib.load("xgboost_fraud_model.pkl")
    prep = joblib.load("preprocessing_pipeline.pkl")
    feat_names = joblib.load("feature_names.pkl")  # list[str]
    return model, prep, feat_names

try:
    model, prep, feature_names = load_artifacts()
except Exception as e:
    st.error(f"Could not load required artifacts: {e}")
    st.stop()

# =============================
# Helpers (combo features + pipeline compatibility)
# =============================
EPS = 1e-5

def compute_combo_features(df: pd.DataFrame) -> pd.DataFrame:
    g = df.copy()
    needed = [
        "nav_time_to_transfers","new_beneficiary","device_changed","sim_changed",
        "tap_dur_mean","backspace_count","swipe_speed_mean","swipe_curviness",
        "press_mean","hour_of_day","travel_speed_kmh","city_changed",
        "hold_std","gap_std","nav_total_time","transfer_amount","roaming",
        "accuracy_m","nav_clicks","nav_screens","tap_count","shortcut_used",
        "amount_z_user"
    ]
    for c in needed:
        if c not in g.columns:
            g[c] = np.nan

    g["fast_new_transfer"] = ((g["nav_time_to_transfers"] < 5) & (g["new_beneficiary"] == 1)).astype(float)
    g["device_sim_change"] = ((g["device_changed"] == 1) & (g["sim_changed"] == 1)).astype(float)
    g["tap_backspace_combo"] = g["tap_dur_mean"].fillna(0) * g["backspace_count"].fillna(0)
    g["speed_curviness_ratio"] = g["swipe_speed_mean"].fillna(0) / (g["swipe_curviness"].fillna(0) + EPS)
    g["pressure_speed_mismatch"] = g["press_mean"].fillna(0) / (g["swipe_speed_mean"].fillna(0) + EPS)
    night_flag = (g["hour_of_day"] < 6).astype(float)
    g["transfer_hour_risk"] = g["transfer_amount"].fillna(0) * night_flag
    g["location_jump_speed"] = ((g["travel_speed_kmh"] > 500) & (g["city_changed"] == 1)).astype(float)
    g["low_variability_typing"] = ((g["hold_std"] < 0.02) & (g["gap_std"] < 0.02)).astype(float)
    g["fast_session_high_value"] = ((g["nav_total_time"] < 10) & (g["transfer_amount"] > 30000)).astype(float)
    g["network_context_risk"] = ((g["roaming"] == 1) & (g["accuracy_m"] > 1000)).astype(float)
    g["frequent_backtrack"] = g["nav_clicks"].fillna(0) / (g["nav_screens"].fillna(0) + EPS)
    g["high_pressure_low_taps"] = ((g["press_mean"] > 1.5) & (g["tap_count"] < 5)).astype(float)
    g["shortcut_absent_slow_nav"] = ((g["shortcut_used"] == 0) & (g["nav_time_to_transfers"] > 20)).astype(float)
    g["amount_zscore_extreme"] = (g["amount_z_user"].abs() > 3).astype(float)
    for c in ["sim_changed","device_changed","city_changed","roaming"]:
        if c not in g.columns:
            g[c] = 0.0
    g["contextal_anomaly_score"] = g[["sim_changed","device_changed","city_changed","roaming"]].fillna(0).sum(axis=1)

    return g

def expected_input_columns(prep) -> tuple[list, list]:
    num_cols, cat_cols = [], []
    for name, trans, cols in prep.transformers_:
        if name == "num":
            num_cols = list(cols)
        elif name == "cat":
            cat_cols = list(cols)
    return num_cols, cat_cols

def ensure_expected_columns(df: pd.DataFrame, prep) -> pd.DataFrame:
    num_cols, cat_cols = expected_input_columns(prep)
    out = df.copy()
    for c in num_cols:
        if c not in out.columns:
            out[c] = np.nan
    for c in cat_cols:
        if c not in out.columns:
            out[c] = "Unknown"
    return out[num_cols + cat_cols]

def predict_fraud_probability(X_proc: np.ndarray) -> np.ndarray:
    prob_legit = model.predict_proba(X_proc)[:, 1]
    return 1.0 - prob_legit  # fraud prob

# =============================
# Sidebar controls (global)
# =============================
with st.sidebar:
    st.header("‚öôÔ∏è Controls")
    threshold = st.slider("Decision threshold (fraud probability)", 0.0, 1.0, 0.50, 0.01)
    st.caption("Lower = catch more fraud (more alerts). Higher = fewer alerts (more conservative).")

# =============================
# Tabs
# =============================
tab_single, tab_batch, tab_insights, tab_about = st.tabs(
    ["üßç Single Session", "üì¶ Batch Scoring", "üîç Feature Insights", "üîß Settings & About"]
)

# =============================
# Tab: Single Session
# =============================
with tab_single:
    st.subheader("Single Session Scoring")
    st.markdown('<div class="section-card">', unsafe_allow_html=True)

    # Grouped inputs
    c1, c2, c3 = st.columns([1,1,1])

    with st.expander("Touch & Swipe", expanded=True):
        col1, col2, col3 = st.columns(3)
        tap_count = col1.number_input("tap_count", value=7.0)
        tap_dur_mean = col1.number_input("tap_dur_mean", value=92.0)
        tap_dur_std = col1.number_input("tap_dur_std", value=18.0)
        press_mean = col2.number_input("press_mean", value=1.2)
        press_std = col2.number_input("press_std", value=0.3)
        swipe_points = col3.number_input("swipe_points", value=12.0)
        swipe_speed_mean = col3.number_input("swipe_speed_mean", value=0.65)
        swipe_speed_std = col3.number_input("swipe_speed_std", value=0.07)
        swipe_curviness = col3.number_input("swipe_curviness", value=0.15)

    with st.expander("Typing", expanded=False):
        col1, col2, col3 = st.columns(3)
        key_count = col1.number_input("key_count", value=23.0)
        hold_mean = col2.number_input("hold_mean", value=105.0)
        hold_std = col2.number_input("hold_std", value=14.0)
        gap_mean = col3.number_input("gap_mean", value=160.0)
        gap_std = col3.number_input("gap_std", value=30.0)
        backspace_count = col1.number_input("backspace_count", value=3.0)

    with st.expander("Navigation", expanded=False):
        col1, col2, col3 = st.columns(3)
        nav_screens = col1.number_input("nav_screens", value=5.0)
        nav_total_time = col2.number_input("nav_total_time", value=22.0)
        nav_clicks = col3.number_input("nav_clicks", value=14.0)
        nav_time_to_transfers = col1.number_input("nav_time_to_transfers", value=6.5)
        shortcut_used = col2.selectbox("shortcut_used", [0, 1], index=0)
        did_open_transfers = col3.selectbox("did_open_transfers", [0, 1], index=1)

    with st.expander("Context & Location", expanded=False):
        col1, col2, col3 = st.columns(3)
        latitude = col1.number_input("latitude", value=-1.286389)
        longitude = col2.number_input("longitude", value=36.817223)
        accuracy_m = col3.number_input("accuracy_m", value=32.0)
        hour_of_day = col1.number_input("hour_of_day", value=1)
        travel_speed_kmh = col2.number_input("travel_speed_kmh", value=0.0)
        city_changed = col3.selectbox("city_changed", [0.0, 1.0], index=0)
        device_changed = col1.selectbox("device_changed", [0.0, 1.0], index=1)
        sim_changed = col2.selectbox("sim_changed", [0.0, 1.0], index=1)
        impossible_speed_flag = col3.selectbox("impossible_speed_flag", [0.0, 1.0], index=0)
        roaming = col1.selectbox("roaming", [0, 1], index=0)

    with st.expander("Payment & User Context", expanded=False):
        col1, col2, col3 = st.columns(3)
        amount_z_user = col1.number_input("amount_z_user", value=2.8)
        transfer_amount = col2.number_input("transfer_amount", value=65000.0)
        new_beneficiary = col3.selectbox("new_beneficiary", [0, 1], index=1)

    with st.expander("Categoricals", expanded=False):
        col1, col2, col3 = st.columns(3)
        day_of_week = col1.selectbox("day_of_week", ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"], index=0)
        city = col2.text_input("city", "Nairobi")
        carrier_name = col3.text_input("carrier_name", "Safaricom")
        network_type = col1.selectbox("network_type", ["2G","3G","4G","5G","WiFi"], index=2)
        wifi_ssid = col2.text_input("wifi_ssid", "Unknown")
        device_model = col3.text_input("device_model", "SM-A546B")
        os_version = col1.text_input("os_version", "Android 13")
        screen_resolution = col2.text_input("screen_resolution", "1080x2400")

    st.markdown("</div>", unsafe_allow_html=True)
    st.write("")

    # Collect to dict and score
    if st.button("Score this session", use_container_width=True):
        row = {
            "tap_count": tap_count, "tap_dur_mean": tap_dur_mean, "tap_dur_std": tap_dur_std,
            "press_mean": press_mean, "press_std": press_std, "swipe_points": swipe_points,
            "swipe_speed_mean": swipe_speed_mean, "swipe_speed_std": swipe_speed_std, "swipe_curviness": swipe_curviness,
            "key_count": key_count, "hold_mean": hold_mean, "hold_std": hold_std, "gap_mean": gap_mean, "gap_std": gap_std,
            "backspace_count": backspace_count, "nav_screens": nav_screens, "nav_total_time": nav_total_time,
            "nav_clicks": nav_clicks, "nav_time_to_transfers": nav_time_to_transfers, "shortcut_used": shortcut_used,
            "did_open_transfers": did_open_transfers, "latitude": latitude, "longitude": longitude, "accuracy_m": accuracy_m,
            "hour_of_day": hour_of_day, "travel_speed_kmh": travel_speed_kmh, "amount_z_user": amount_z_user,
            "transfer_amount": transfer_amount, "city_changed": city_changed, "device_changed": device_changed,
            "sim_changed": sim_changed, "impossible_speed_flag": impossible_speed_flag, "new_beneficiary": new_beneficiary,
            "roaming": roaming, "day_of_week": day_of_week, "city": city, "carrier_name": carrier_name,
            "network_type": network_type, "wifi_ssid": wifi_ssid, "device_model": device_model,
            "os_version": os_version, "screen_resolution": screen_resolution
        }
        one_df = pd.DataFrame([row])
        one_df = compute_combo_features(one_df)
        one_df = ensure_expected_columns(one_df, prep)
        X_one = prep.transform(one_df)
        prob = predict_fraud_probability(X_one)[0]
        alert = prob >= threshold

        k1, k2, k3 = st.columns(3)
        k1.metric("Fraud Probability", f"{prob:.3f}")
        k2.metric("Decision Threshold", f"{threshold:.2f}")
        k3.metric("Alert", "YES" if alert else "NO")

# =============================
# Tab: Batch Scoring
# =============================
with tab_batch:
    st.subheader("Batch Scoring (CSV)")
    st.caption("Upload raw sessions (same column names as training). Include `label` (0=fraud, 1=legit) for metrics.")

    # Template CSV
    def build_template_csv() -> bytes:
        num_cols, cat_cols = expected_input_columns(prep)
        # Remove engineered columns from template (they‚Äôre computed on the fly)
        engineered = {
            "fast_new_transfer","device_sim_change","tap_backspace_combo","speed_curviness_ratio",
            "pressure_speed_mismatch","transfer_hour_risk","location_jump_speed","low_variability_typing",
            "fast_session_high_value","network_context_risk","frequent_backtrack","high_pressure_low_taps",
            "shortcut_absent_slow_nav","amount_zscore_extreme","contextal_anomaly_score"
        }
        cols = [c for c in num_cols + cat_cols if c not in engineered]
        example = pd.DataFrame([{c: "" for c in cols}])
        return example.to_csv(index=False).encode("utf-8")

    with st.expander("Need a template CSV? Click to download", expanded=False):
        st.download_button("Download template", build_template_csv(), "template_raw_sessions.csv", "text/csv")

    uploaded = st.file_uploader("Upload CSV", type=["csv"])
    if uploaded:
        df_raw = pd.read_csv(uploaded)
        st.write("Preview:", df_raw.head())

        df_ready = compute_combo_features(df_raw)
        df_ready = ensure_expected_columns(df_ready, prep)

        X = prep.transform(df_ready)
        prob = predict_fraud_probability(X)
        alerts = (prob >= threshold).astype(int)

        scored = df_raw.copy()
        scored["fraud_probability"] = prob
        scored["fraud_alert"] = alerts

        # Metrics if labels exist
        if "label" in df_raw.columns:
            y_true = df_raw["label"].values  # 0=fraud, 1=legit
            y_fraud = (y_true == 0).astype(int)
            prec = precision_score(y_fraud, alerts, zero_division=0)
            rec  = recall_score(y_fraud, alerts, zero_division=0)
            f1   = f1_score(y_fraud, alerts, zero_division=0)
            auc  = roc_auc_score(y_fraud, prob)
            y_pred_label = (1 - alerts)
            acc = accuracy_score(y_true, y_pred_label)

            m1, m2, m3, m4, m5 = st.columns(5)
            m1.metric("Accuracy", f"{acc:.3f}")
            m2.metric("Precision (fraud)", f"{prec:.3f}")
            m3.metric("Recall (fraud)", f"{rec:.3f}")
            m4.metric("F1 (fraud)", f"{f1:.3f}")
            m5.metric("ROC-AUC", f"{auc:.3f}")

            # Confusion Matrix
            cm = confusion_matrix(y_fraud, alerts)
            fig_cm, ax_cm = plt.subplots(figsize=(4,4))
            ConfusionMatrixDisplay(cm, display_labels=["Legit","Fraud"]).plot(ax=ax_cm, cmap="Blues", colorbar=False)
            ax_cm.set_title("Confusion Matrix (thresholded)")
            st.pyplot(fig_cm)

            # ROC
            fpr, tpr, thr = roc_curve(y_fraud, prob)
            fig_roc, ax_roc = plt.subplots(figsize=(5,4))
            ax_roc.plot(fpr, tpr, lw=2)
            ax_roc.plot([0,1], [0,1], linestyle="--")
            ax_roc.set_xlabel("False Positive Rate")
            ax_roc.set_ylabel("True Positive Rate (Recall)")
            ax_roc.set_title("ROC Curve")
            st.pyplot(fig_roc)

        st.download_button(
            "‚¨áÔ∏è Download scored CSV",
            scored.to_csv(index=False).encode("utf-8"),
            "scored_predictions.csv",
            "text/csv"
        )

# =============================
# Tab: Feature Insights
# =============================
with tab_insights:
    st.subheader("Top Feature Importances")
    try:
        importances = getattr(model, "feature_importances_", None)
        if importances is None or len(importances) == 0:
            st.info("Feature importances are not available for this model object.")
        else:
            n = min(len(importances), len(feature_names))
            imp_df = pd.DataFrame({"feature": feature_names[:n], "importance": importances[:n]})
            topk = imp_df.sort_values("importance", ascending=False).head(20)

            fig_imp, ax_imp = plt.subplots(figsize=(8,6))
            ax_imp.barh(topk["feature"][::-1], topk["importance"][::-1])
            ax_imp.set_title("Top 20 Features")
            ax_imp.set_xlabel("Importance")
            st.pyplot(fig_imp)

            st.dataframe(topk.reset_index(drop=True))
    except Exception as e:
        st.info(f"Could not display feature importances: {e}")

# =============================
# Tab: Settings & About
# =============================
with tab_about:
    st.subheader("Settings & About")
    st.markdown("""
**Label convention:** `1 = legit`, `0 = fraud`
**Decision policy:** An alert is raised when *fraud probability ‚â• threshold*.
**Threshold tips:** Lower the threshold to catch more fraud (higher recall), raise it to reduce false positives.
**Data expectations:** The app computes the same engineered features used during training to keep the model inputs consistent.
    """)
    st.markdown("---")
    st.caption("Built for capstone demonstration ‚Äî behavioral biometrics, mobile banking fraud detection.")


```

## üè¶ Title  
**Behavioral Biometric Security System for Mobile Banking Fraud Detection in Kenya**

## 1) Problem & Motivation
Mobile banking is now essential in Kenya, but **post-login fraud** is rising. Attackers bypass PINs/OTPs/biometrics via **SIM-swap, remote access tools, social engineering, bots, and high-value mule transfers**. Traditional controls **end at login**‚Äîapps cannot verify that the person using the app *after* login is still the genuine user.

## 2) Goal
Detect fraudulent sessions **during usage** using **behavioral biometrics**: touch, swipe, typing, navigation, and context, plus combined (multi-signal) features.

## 3) Data & Preprocessing (Summary)
- Synthetic dataset: **~700 users**, many sessions per user, **~5% fraud**.
- Parsed JSON ‚Üí numeric features; built per-user ‚Äúprevious session‚Äù context (city/SIM/device changes).
- Imputation, one-hot encoding, scaling, **grouped train/test split by user** to prevent leakage.

## 4) Feature Engineering
- **Base features:** touch (tap counts, durations, pressures), typing (hold/gap/backspaces), navigation (screen order, dwell, clicks), context (time, GPS, SIM/device).
- **Combo features (examples):**
  - `device_sim_change` (SIM + device swap)
  - `fast_new_transfer` (quick route to transfers + new beneficiary)
  - `speed_curviness_ratio` (fast, straight swipes ‚Üí bot-like)
  - `amount_zscore_extreme` (amount far from user norm)
  - `contextal_anomaly_score` (sum of unusual context flags)

## 5) Modeling & Evaluation
- Models: **Logistic Regression**, **Random Forest**, **XGBoost**.
- Best: **XGBoost** with tuned hyperparameters (RandomizedSearchCV).
- Metrics (test set): **high recall and AUC**, very low false positives.  
  *Confusion matrix, ROC curve, and feature importances confirm validity.*

## 6) Results (What We Proved)
- **Behavioral patterns ‚â† identity**‚Äîthey reveal *who is actually using the app now*.  
- Even with correct PIN/OTP/biometric, impostors **behave differently**:
  - SIM/device/city jumps
  - Unnatural, rushed navigation
  - Robotic typing/swipes
  - Out-of-profile amounts at odd hours
- Combining base + combo features yields **strong, explainable detection**.

## 7) Deployment Demo
A **Streamlit dashboard**:
- **Single session scoring** (manual inputs)
- **Batch CSV scoring** + metrics (Accuracy, Precision, Recall, F1, ROC-AUC)
- **Threshold slider** for risk appetite
- **Top feature importances** for explainability

## 8) Conclusion
This project closes the **post-login security gap** by continuously verifying user behavior. It shows that **behavioral biometrics** can detect SIM-swap, RAT, social-engineering, bot-like, and high-value outlier fraud **in real time**‚Äî**before money is lost**.

**Key takeaway:** Instead of asking *‚ÄúWho logged in?‚Äù*, banks can now ask  
**‚ÄúIs this still the real user using the app right now?‚Äù**

## 9) Limitations & Future Work
- Synthetic data: ethical and necessary; future work includes **pilot with real data** under privacy-preserving agreements.
- **MLOps**: production drift monitoring, feedback loops, periodic retraining.
